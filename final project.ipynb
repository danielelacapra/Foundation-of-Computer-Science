```python
#Distribute the revenue according to the order of appearance in a movie. 
#Assume that the i-th actor contributes twice as much as the (i+1)-th actor to the revenue.
import numpy as np
def rev_actor (orders,number_actors, revenue):
    part = [0]*len(orders)
    for i in range(len(orders)):
        k= orders[i]
        n= number_actors[i]
        reve=revenue[i]
        e_n = np.exp2(n)
        e_k = np.exp2(k)
        
        number = reve/e_k + reve/(e_k * (e_n -1))
        part[i]= np.round(number/2)
    return part
```
